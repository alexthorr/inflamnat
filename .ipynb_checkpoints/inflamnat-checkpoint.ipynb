```python
import pandas as pd

# Load your dataset (replace with your actual file path)
df = pd.read_csv("InflamNat_subset.csv")

# Ensure the column exists
if "np_classifier_class" not in df.columns:
    raise ValueError("Column 'np_classifier_class' not found in the dataset.")

# Find the top 5 most frequent classes
top_5_classes = df["np_classifier_class"].value_counts().nlargest(5).index

# Filter the dataframe to include only those classes
df_top5 = df[df["np_classifier_class"].isin(top_5_classes)].copy()

# Display basic info
print("Top 5 classes:", list(top_5_classes))
print(f"Filtered dataframe shape: {df_top5.shape}")

# (Optional) Save the filtered dataframe
df_top5.to_csv("top5_np_classifier_class.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load your data ---
admet = pd.read_excel("top5_np_classifier_class.xlsx", sheet_name='admetlab')

# --- List of ADMET columns to plot ---
admet_columns = [
    "OATP1B1", "OATP1B3", "BCRP", "BSEP", "MRP1", "PPB", "logVDss", "Fu"
]

# --- Create boxplots for each ADMET column ---
for col in admet_columns:
    if col not in admet.columns:
        print(f"⚠️ Column '{col}' not found in DataFrame, skipping.")
        continue
    
    plt.figure(figsize=(10,6))
    sns.boxplot(
        data=admet,
        x="np_classifier_class",
        y=col,
        palette="Set2",
        showfliers=False
    )
    plt.title(f"{col} distribution by np_classifier_class", fontsize=14)
    plt.xlabel("np_classifier_class", fontsize=12)
    plt.ylabel(col, fontsize=12)
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```
