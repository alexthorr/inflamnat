```python
import numpy as np
vars_heatmap = ["MW", "Vol", "nHA", "nHD", "TPSA", "nRot", "logD", "logP"]

# Filter only selected variables
df_sub = prop[vars_heatmap]

# ----------------------------
# Compute correlation matrix
# ----------------------------
corr = df_sub.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))

# ----------------------------
# Plot heatmap
# ----------------------------
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", square=True, cmap="coolwarm", mask=mask)
plt.title("Correlation Heatmap of Molecular Descriptors")
plt.tight_layout()
plt.show()
```

# Análisis descriptivo

```python
# Global plot style

sns.set_theme(style="whitegrid", context="talk")

# Load data
df = pd.read_excel('top5_np_classifier_class.xlsx')

# Columns to analyze
cols = ["MW", "Vol", "nHA", "nHD", "TPSA", "logS", "logD"]
x_col = "np_classifier_class"

# 1. BASIC INFO
print("\n=== BASIC INFO ===")
print(df[[x_col] + cols].info())


# 2. SUMMARY STATS BY CLASS
print("\n=== SUMMARY STATISTICS BY CLASS ===")
print(df.groupby(x_col)[cols].describe())


# 3. COUNT PLOT (class distribution)
plt.figure(figsize=(10,6))
sns.countplot(x=x_col, data=df, palette="Set2")
plt.title("Class Distribution")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. BOXPLOTS (x = class, y = descriptor)
plt.figure(figsize=(18, 20))

for i, col in enumerate(cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(data=df, x=x_col, y=col, palette="Set3")
    plt.title(f"Boxplot of {col} by Class")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 5. VIOLINPLOTS (optional, nicer distributions)
plt.figure(figsize=(18, 20))

for i, col in enumerate(cols, 1):
    plt.subplot(4, 2, i)
    sns.violinplot(data=df, x=x_col, y=col, palette="Pastel1", cut=0)
    plt.title(f"Violinplot of {col} by Class")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 6. STRIP + BOX (hybrid plot)
plt.figure(figsize=(18, 20))

for i, col in enumerate(cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(data=df, x=x_col, y=col, color="white", showcaps=True)
    sns.stripplot(data=df, x=x_col, y=col, color="black", alpha=0.4)
    plt.title(f"{col} Distribution per Class")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# 7. CORRELATIONS PER CLASS (optional)
for cls in df[x_col].unique():
    subset = df[df[x_col] == cls][cols]
    plt.figure(figsize=(8, 6))
    sns.heatmap(subset.corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title(f"Correlation Heatmap — Class: {cls}")
    plt.tight_layout()
    plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Directory to save plots
save_dir = "plots_EDA"
os.makedirs(save_dir, exist_ok=True)

# -------------------------------
# SAVE BOXPLOTS
# -------------------------------
plt.figure(figsize=(18, 20))

for i, col in enumerate(cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(data=df, x=x_col, y=col, palette="Set3")
    plt.title(f"Boxplot of {col} by Class")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig(os.path.join(save_dir, "boxplots_by_class.png"), dpi=300)
plt.close()


# -------------------------------
# SAVE VIOLINPLOTS
# -------------------------------
plt.figure(figsize=(18, 20))

for i, col in enumerate(cols, 1):
    plt.subplot(4, 2, i)
    sns.violinplot(data=df, x=x_col, y=col, palette="Pastel1", cut=0)
    plt.title(f"Violinplot of {col} by Class")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig(os.path.join(save_dir, "violinplots_by_class.png"), dpi=300)
plt.close()


# -------------------------------
# SAVE STRIP + BOXPLOT
# -------------------------------
plt.figure(figsize=(18, 20))

for i, col in enumerate(cols, 1):
    plt.subplot(4, 2, i)
    sns.boxplot(data=df, x=x_col, y=col, color="white")
    sns.stripplot(data=df, x=x_col, y=col, color="black", alpha=0.4)
    plt.title(f"{col} Distribution by Class")
    plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig(os.path.join(save_dir, "strip_boxplots_by_class.png"), dpi=300)
plt.close()


# -------------------------------
# SAVE CORRELATION HEATMAP PER CLASS
# -------------------------------
for cls in df[x_col].unique():
    subset = df[df[x_col] == cls][cols]

    plt.figure(figsize=(8, 6))
    sns.heatmap(subset.corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title(f"Correlation Heatmap — Class: {cls}")
    plt.tight_layout()
    plt.savefig(os.path.join(save_dir, f"correlation_heatmap_{cls}.png"), dpi=300)
    plt.close()
```
