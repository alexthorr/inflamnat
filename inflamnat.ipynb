```python
import pandas as pd

# --- 1️⃣ Load dataset ---
# Replace with your actual filename, e.g., "coconut-09-2024.csv"
df = pd.read_csv("coconut_csv-11-2025.csv", encoding="utf-8")

# --- 2️⃣ Filter rows where 'collections' contains 'InflamNat' (case-insensitive) ---
filtered_df = df[df['collections'].astype(str).str.contains('InflamNat', case=False, na=False)]

# --- 3️⃣ Save the filtered data ---
filtered_df.to_csv("InflamNat_subset.csv", index=False, encoding="utf-8")

# --- 4️⃣ Show basic info ---
print(f"✅ Filas totales: {len(df)}")
print(f"✅ Filas filtradas (InflamNat): {len(filtered_df)}")
print("Archivo guardado como 'InflamNat_subset.csv'")
```
